<p><b># Ejercicio 6:</b> Convierte los valores de la columna 'Plays' de 'games.csv' del formato miles(k) a formato numerico.<p>

```python
def convert_k_to_number(knum):
    if 'K' in str(knum):
        knum = float(str(knum).replace('K', '')) * 1000
        return knum
    else:
        return knum

datos['Plays'] = datos['Plays'].apply(convert_k_to_number)

datos.head()
```
